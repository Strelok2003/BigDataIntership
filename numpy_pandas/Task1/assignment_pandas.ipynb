**11. Group people into age groups *young*, *adult*, *retiree*, where:**
* *young* corresponds to 16-35 years old
* *adult* - 35-70 years old
* *retiree* - 70-100 years old

**Assign the name of the corresponding group to each person in a new column called 'AgeGroup'.**

```python
# your code
conditions = [
    (data['age'] >= 16) & (data['age'] < 35),
    (data['age'] >= 35) & (data['age'] < 70),
    (data['age'] >= 70) & (data['age'] <= 100)
]

choices = ['young', 'adult', 'retiree']

data['AgeGroup'] = np.select(conditions, choices, default=None)
data.head()
```

**12-13. Determine the number of people earning >50K in each of the age groups (AgeGroup column), and also output the name of the age group that most frequently earns more than 50K (>50K).**

```python
# your code
conditions = [
    (data['age'] >= 16) & (data['age'] < 35),
    (data['age'] >= 35) & (data['age'] < 70),
    (data['age'] >= 70) & (data['age'] <= 100)
]

choices = ['young', 'adult', 'retiree']

data['AgeGroup'] = np.select(conditions, choices, default=None)

data[data["salary"] == ">50K"].groupby("AgeGroup").size().reset_index(name="count").sort_values(by="count", ascending=False)
```

**14. Group people by occupation (occupation column) and determine the number of people in each group. Then, write a filtering function *filter_func* that will return only those groups where the average age (age column) is no more than 40 and where all workers work more than 5 hours per week (hours-per-week column).**

```python
# your code
def filter_func(group):
    return (group["age"].mean() <= 40) and (group["hours-per-week"].min() > 5)

result = (
    data.groupby("occupation")
    .filter(filter_func)
    .groupby("occupation")
    .size()
    .reset_index(name="count")
)
result
```
